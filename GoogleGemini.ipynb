```python
import google.generativeai as genai
import pathlib
import textwrap
from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
    text = text.replace('•',' *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

#Example usage:
input_text = "This is a • sample text with bullet points"
result = to_markdown(input_text)

display(result)
```

*Chat conversations*

Gemini enables you to have freeform conversations across multiple turns. The ChatSession class simplifies the process by managing the state of the conversation, so unlike with generate_content, you do not have to store the conversation history as a list.

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
story = model.generate_content("Tell me the story about the avengers?").text
to_markdown(story)
```

```python
response = model.generate_content(
    "tell me the story about avengers?",
    generation_config=genai.types.GenerationConfig(
        candidate_count=1,
        stop_sequences=["x"],
        max_output_tokens=500000000,
        temperature=1.0
    )
)

#to_markdown(response)
```

```python
text = response.text

if response.candidates[0].finish_reason.name == "MAX_TOKENS":
    text += '...'

to_markdown(text)
```

```python
to_markdown(response.text)
```

```python
chat=model.start_chat(history=[])
```

```python
response =  chat.send_message("In one sentence, what is halloween?")
```

```python
to_markdown(response.text)
```

```python
chat.history
```

```python
response = chat.send_message("Okay, please explain about halloween to me in detail?",stream=True)
```

```python
for chunk in response:
    print(chunk.text)
    print("_"*80)
```

```python
chat.history
```

```python
for message in chat.history:
    display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```
